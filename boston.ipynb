```python
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/maneeshd/boston-housing/master/housing.csv')
```

```python
df.head()
```

```python
f, (ax1, ax2, ax3) = plt.subplots(1, 3, sharey=True)
f.set_figwidth(15)
ax1.scatter(df['RM'], df['MEDV'])
ax1.set_title('Rooms')
ax2.scatter(df['LSTAT'], df['MEDV'])
ax2.set_title('Lower Status Pop')
ax3.scatter(df['PTRATIO'], df['MEDV'])
ax3.set_title('Student Teacher Ratio')
plt.tight_layout()
plt.show();
```

```python
# df.plot(x='RM', y='MEDV', kind='scatter')
# df.plot(x='PTRATIO', y='MEDV', kind='scatter')
# plt.show();
f, ax1 = plt.subplots(1, 1, sharey=True)
ax1.scatter(df['LSTAT'], df['MEDV'])
ax1.scatter(df['RM'], df['MEDV'])
plt.show();
```

1. Create a dataframe object `X` that is the RM, LSTAT, and PTRATIO columns
2. Create a series object `y` that is the MEDV column pls pls

```python
X = df[['RM', 'LSTAT', 'PTRATIO']]
y = df['MEDV']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
lr = LinearRegression(normalize=True, n_jobs=4)
```

```python
lr.fit(X_train, y_train)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
X_train.columns
```

```python
lr.score(X_test, y_test)
```

```python
two_houses = [
    [5, 33, 18],
    [8, 5, 10]
]
```
